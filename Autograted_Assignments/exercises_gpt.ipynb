```python

y = np.linspace(-5,8,400)

# Define uncoditional probability density function of y
f = lambda y: 0.3 *((1/(np.sqrt(2*np.pi))) * np.exp(-y**2 / 2)) + 0.7 * ((1/(np.sqrt(2*np.pi))) * np.exp(-((y-3)**2) / 2))

#Plot
plt.figure(figsize=(8,5))
plt.plot(y, f(y), color='black', linewidth=2, label='Mixture pdf f_Y(y)')
plt.title("Mixture of Two Normal Distributions")
plt.xlabel("y")
plt.ylabel("Density f_Y(y)")
plt.legend()
plt.grid(True, linestyle=':')
plt.show()
```

```python
x = np.linspace(-1,3,400)
f = np.where((x >= 0) & (x <= 2),x / 2, 0)       # pdf of X
mean_x = 8/6
#Plot
plt.figure(figsize=(5,5))
plt.plot(x, f, color='black', linewidth=2, label='PDF f_X(x)')
# Add the mean line
plt.axvline(mean_x, color='red', linestyle='--', linewidth=2, label=f'Mean E[X]={mean_x:.2f}')
plt.title("Distribution")
plt.xlabel("x")
plt.ylabel("Density f_X(x)")
plt.legend()
plt.grid(True, linestyle=':')
plt.show()
```

```python
#Exercise 1/CLT
from scipy.stats import norm
# Parameters
n = 30
samples = 1000
mu, sigma2 = 0.5, 1/12      # expected value and variance
sigma = np.sqrt(sigma2/n)   # standard deviation for each sample
# Simulate 1000 sample means
rng = np.random.default_rng(42)
sample_means = [rng.uniform(0, 1, n).mean() for _ in range(samples)]
print(sample_means)
# Plot histogram
xbar = sample_means
plt.hist(xbar, bins=30, density=True, edgecolor = "black", alpha = 0.7)
# Overlay theoretical normal curve from CLT
x = np.linspace(min(sample_means), max(sample_means), 200)
plt.plot(x, norm.pdf(x, mu, sigma), 'r-', lw=2, label='Normal Approximation (CLT)')
plt.title("Distribution of sample means")
plt.xlabel("Sample mean values")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Exercise 3/
# Parameters
n = 30
samples = 1000
p = 0.3
# Simulate 1000 sample means and calculate MLE for each sample
sample_MLEs = np.empty(samples)
for s in range(samples):
    sample = np.random.binomial(1, p, n)     # one sample of size n
    sample_MLEs[s] = sample.mean()              # add MLE for that sample to the MLEs array
print(f"Average MLE over {samples} is: {sample_MLEs.mean()} which is close to {p}")
```
